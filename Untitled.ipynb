```python
# clean spaces first (VERY IMPORTANT)
data['service_date'] = data['service_date'].str.strip()

data['service_day'] = data['service_date'].str[:2].astype(int)
data['service_month'] = data['service_date'].str[3:5].astype(int)
data['service_year'] = data['service_date'].str[6:10].astype(int)
```

```python
data['interaction_date'] = data['interaction_date'].str.strip()

data['interaction_day'] = data['interaction_date'].str[:2].astype(int)
data['interaction_month'] = data['interaction_date'].str[3:5].astype(int)
data['interaction_year'] = data['interaction_date'].str[6:10].astype(int)
```

```python
data['service_key'] = data['service_key'].str.strip()
data['service_year_key'] = data['service_key'].str[:4].astype(int)
data['service_month_key'] = data['service_key'].str[5:7].astype(int)
data['service_day_key'] = data['service_key'].str[8:10].astype(int)
```
